```python
def plotEtaSum(mtx):
    figure(figsize=(10,10))
    eta = [int(re.search('eta(.+?)-phi', m).group(1)) for m in mtx.index]
    count = np.count_nonzero(mtx, axis=1)
    b, slp = polyfit(eta, count, 1)
    plt.plot(eta, count, '.')
    plt.plot(eta, b + slp * np.array(eta), '-')
    plt.xticks(np.arange(0,22, 2))
    plt.xlabel('Eta', fontsize=18)
    plt.ylabel('Number of sums', fontsize=18)
    plt.tick_params(axis='both', which='major', labelsize=18)
```

```python
plotEtaSum(mtx_em)
```
